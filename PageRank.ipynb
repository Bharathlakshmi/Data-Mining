```python
import networkx as nx
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import operator
import random as rd
```

```python
# create a directed graph
graph=nx.gnp_random_graph(8,0.6,directed=True)

#draw a graph
nx.draw(graph,with_labels=True,font_color='blue',font_size=15,node_color='pink')

#plot a graph
plt.show()
```

```python

count=graph.number_of_nodes()
print(count)
```

```python
#graph neighbours of a node 5
print(list(graph.neighbors(5)))
```

```python
rank_dict={}
x=rd.randint(0,8)
for j in range(0,8):
  rank_dict[j]=0
rank_dict[x]=rank_dict[x]+1
for i in range(600000):
  list_n=list(graph.neighbors(x))
  if(len(list_n)==0):
    x=rd.randint(0,8)
    rank_dict[x]=rank_dict[x]+1
  else:
    x=rd.choice(list_n)
    rank_dict[x]=rank_dict[x]+1
print("Random Walk Score Updated")
```

```python
#normalising values
for j in range(0,8):
  rank_dict[j]=rank_dict[j]/600000

#Page rank by networkx library
pagerank=nx.pagerank(graph)

#sorting both dictionaries based on items
pagerank_sorted=sorted(pagerank.items(),key=lambda v:(v[1],v[0]),reverse=True)

pagerank_sorted

#sorting the rank_dict based on values
rank_dict_sorted=sorted(rank_dict.items(),key=lambda v:(v[1],v[0]),reverse=True)

rank_dict_sorted
```

```python
print("Order generated by the implementation algorithm :\n")
for i in rank_dict_sorted:
  print(i[0],end=" ")
```
